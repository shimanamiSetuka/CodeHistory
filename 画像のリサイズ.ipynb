```python
import cv2
import math


if __name__ == "__main__":
  img = cv2.imread('./0810_01_A.tif')

  color = (99, 215, 163)

  height, width, color = img.shape # 画像の縦横サイズを取得

  # 縦長画像→幅を拡張する
  if height > width:
    diffsize = height - width
    # 元画像を中央ぞろえにしたいので、左右に均等に余白を入れる
    padding_half = int(diffsize / 2)
    color = (99,215,163)
    padding_img = cv2.copyMakeBorder(img, 0, 0, padding_half, padding_half, cv2.BORDER_CONSTANT, value=color)
    cv2.imwrite('./example1@padding.tif', padding_img)

  # 横長画像→高さを拡張する
  elif width > height:
    diffsize = width - height
    padding_half = int(diffsize / 2)
    color = (99,215,163)
    padding_img = cv2.copyMakeBorder(img, padding_half, padding_half, 0, 0, cv2.BORDER_CONSTANT, value=color)
    cv2.imwrite('./0810_01_A@padding.tif', padding_img)
```

```python
from pathlib import Path

import cv2
import numpy as np

img = cv2.imread("/content/drive/MyDrive/PBL_data/resize_file2/padding/0808_00_Apadding.jpg")

size = (320, 320)  # 分割後の大きさ
rows = int(np.ceil(img.shape[0] / size[0]))  # 行数
cols = int(np.ceil(img.shape[1] / size[1]))  # 列数

chunks = []
for row_img in np.array_split(img, rows, axis=0):
    for chunk in np.array_split(row_img, cols, axis=1):
        chunks.append(chunk)
print(len(chunks))

# 保存する。
output_dir = Path("output1")
output_dir.mkdir(exist_ok=True)
for i, chunk in enumerate(chunks):
    save_path = output_dir / f"chunk_{i:01d}.jpg"
    cv2.imwrite(str(save_path), chunk)
```

このようなExcelファイルからデータを取得する

```python
from openpyxl import load_workbook
from PIL import Image


# Excelファイルのロード(読み取り専用)
excel_path='/content/drive/MyDrive/PBL_data/azamiuma.xlsx'
workbook = load_workbook(filename=excel_path, read_only=True)
# Excelのシート名一覧を表示
print(workbook.sheetnames)
worksheet = workbook['Sheet1']

for row in worksheet.iter_rows(min_row=2):
    print(f"[ {row[0].row}行目 ]")
    # 行からセルを1個ずつ取得し、処理をする
    for cell in row:
        print(f"{cell.column}列目：{cell.value}")
        
        if cell.column == 1:
         file_path = '/content/drive/MyDrive/PBL_data/JPG/' + cell.value
         file_path = file_path + '.jpg'
         im = Image.open(file_path)

         change_path = '/content/drive/MyDrive/PBL_data/resize_file2/' + cell.value

# 列ごとに値を保存する

        elif cell.column == 2:
          x = cell.value

        elif cell.column == 3:
          y = cell.value

        elif cell.column == 4:
          w = cell.value

        elif cell.column == 5:
          h = cell.value

        elif cell.column == 6:
          if cell.value == 'チャノキイロアザミウマ':
            tekito = 'C'

          elif cell.value == '別種アザミウマ':
            tekito = 'D'

    change_path = change_path + tekito
    change_path = change_path +'.jpg'
    im_crop = im.crop((x, y, x + w, y + h))
    im_crop.save(change_path, quality=95)
     
    print('------------------------------------------')

# ロードしたExcelファイルを閉じる
workbook.close()
```

```python
import glob
import random
import os 
import shutil
import math

INPUT_DIR = '/content/drive/MyDrive/PBL_data/resize_file'
OUTPUT_DIR = '/content/drive/MyDrive/PBL_data/valid_data'
#ランダムで抽出する割合
SAMPLING_RATIO = 0.5

def random_sample_file():
    files = glob.glob(INPUT_DIR + '/*.tif')

    random_sample_file = random.sample(files,math.ceil(len(files)*SAMPLING_RATIO))
    os.makedirs(OUTPUT_DIR + "/choices",exist_ok=True)

    for file in random_sample_file:
        shutil.copy2(file,OUTPUT_DIR + "/choices/")

if __name__ == '__main__':
    random_sample_file()
```

```python
from keras.preprocessing.image import load_img, img_to_array
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
 
# 入力ディレクトリを作成
input_dir = "/content/drive/MyDrive/PBL_data/resize_file2/test"
files = glob.glob(input_dir + '/*.jpg')
 
# 出力ディレクトリを作成
output_dir = "/content/drive/MyDrive/PBL_data/resize_file2/new_test"
if os.path.isdir(output_dir) == False:
    os.mkdir(output_dir)
 
 
for i, file in enumerate(files):
 
    img = load_img(file)
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
 
    # ImageDataGeneratorの生成
    datagen = ImageDataGenerator(
        height_shift_range=0.5
    )
 
    # 1個の画像を生成します
    g = datagen.flow(x, batch_size=1, save_to_dir=output_dir, save_prefix='img', save_format='jpg')
    for i in range(1):
        batch = g.next()
```
